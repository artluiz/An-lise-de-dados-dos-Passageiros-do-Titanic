```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_df = pd.read_csv('titanic.csv')

survived_count = titanic_df['Survived'].value_counts()
plt.bar(survived_count.index, survived_count.values)
plt.xticks([0, 1], ['Não Sobreviveu', 'Sobreviveu'])
plt.ylabel('Número de Passageiros')
plt.title('Sobreviventes do Titanic')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_df = pd.read_csv('titanic.csv')

gs = titanic_df.groupby('Sex')

for sex, group in gs:
    
    survived_count = group['Survived'].value_counts()
    
    plt.bar(survived_count.index, survived_count.values)
    plt.xticks([0, 1], ['Não Sobreviveu', 'Sobreviveu'])
    plt.ylabel('Número de Passageiros')
    plt.title(f'Sobreviventes do Titanic ({sex})')
    plt.show()
```
